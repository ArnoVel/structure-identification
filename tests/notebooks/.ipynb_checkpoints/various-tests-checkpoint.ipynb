```python
import torch
import numpy as np
from scipy.stats import gamma
from ..functions.kernels import RBF,RQ, SumIdentical, SumKernels
from ..functions.operations import block_matrix, cross_average, unblock_matrix
from itertools import product
```

```python
from dependence.mmd import MMD, numpy, DEVICE
```

```python
params = [{'bandwidth':10**(i),
            'alpha':k
            } for i,k in zip(range(-2,2),range(1,6))]
kernels = [RQ for p in params]
kerlist = [ker(**p, precompute=True) for (ker,p) in product(kernels,params)]

paramlist = list(product(kernels,params))
```

```python
import matplotlib.pyplot as plt
```

```python
EPS = torch.linspace(0,40,10000)
NEPS = torch.linspace(-40,0,10000)
X = EPS.to(DEVICE).unsqueeze(1)
negX = NEPS.to(DEVICE).unsqueeze(1)
```

```python
from functions.kernels import PairwiseSqDist
```

```python
sqDists = PairwiseSqDist(X)
D = sqDists.sqrt()
Dneg = PairwiseSqDist(negX).sqrt()
```

```python
D.min(), sqDists.min()
```

```python
print(D.max())
for k in kerlist:
    print(k(D.max()), k(D).max())
```

```python
for i,k in enumerate(kerlist):
    Y = k(D)[:,0]
    negY = k(Dneg)[:,-1]
    plt.plot(torch.cat([NEPS,EPS]),
             numpy(torch.cat([negY,Y])), label=str(paramlist[i]))
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```
